```python
def log_base(n,base= 5):
    # Devuelve el logaritmo de un número y la base
    return np.log(n) / np.log(base)
```

```python
log_base(8,2)
```

```python
# Usar con los datos de presión

df_logPressure = log_base(df['Pressure'],10)
```

```python
# Aplicar la función log base a nuestro dataframe:
# Opción 1 - Función anónima lambda para feature engineering

df['Pressure'].apply( lambda x : np.log(x) )
```

```python
# Opción 2 - Función nombrada

def feature_engineering_presion():
    presiones = df['Pressure'].to_list()
    res = []
    for i in presiones:
        res.append(np.log(i))
    return res

feature_engineering_presion()
```

```python
# Opción 3 - Ejecución directa entre numpy y una serie de pandas

np.log(df['Pressure']) / np.log(5)
```

```python
# Raiz cuadrada de Permeabilidad

np.sqrt(df['Permeability'])
```

```python
# Euler a la n (Exp(Porosity))

np.exp(df['Porosity'])
```

```python
df['log_pressure'] = df['Pressure'].apply( lambda x : np.log(x) )
df['sqrt_permeability'] = df['Permeability'].apply( lambda x : np.sqrt(x) )
df['exp_porosity'] = df['Porosity'].apply( lambda x : np.exp(x) )


#df = df.drop(columns=['log_pressure'])
#df = df.drop(columns=['sqrt_permeability'])
#df = df.drop(columns=['exp_porosity'])
```

```python
df
```

```python
# Transformaciones Multivariables

df['Productivity_Index'] = df['FlowRate']/((df['Pressure']-df['BottomHolePressure']))
```

```python
df['Reynolds_Index'] = df['FlowRate'] * df['Density'] / df['Viscosity']
```

```python
df['Porosity_Thickness_Index'] = df['Porosity'] / df['ReservoirThickness']
```

```python
df
```

```python
# Feature Engineering con métodos empíricos
# Relación de variables en un polinomio

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree= 3)
```

```python
features = poly.fit_transform(df[['FlowRate','Density','Viscosity']])
```

```python
reynolds_poly_df = pd.DataFrame(features)
```

```python
pd.concat([df,reynolds_poly_df], axis= 1)
```
